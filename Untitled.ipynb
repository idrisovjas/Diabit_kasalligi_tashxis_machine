## ML

```python
from sklearn.preprocessing import StandardScaler
sk = StandardScaler()
X = df.drop('Outcome',axis=1).values
y = df['Outcome']

X = sk.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split 

X_train , X_test , y_train , y_test = train_test_split( X ,y , test_size = 0.2 , random_state = 12)
```

```python
X
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=11)
```

```python
knn.fit(X_train , y_train )
```

```python
y_predict = knn.predict(X_test)
```

## Baholash Jaccard Score  

```python
from sklearn.metrics import jaccard_score
jaccard_score(y_test , y_predict)
```

### Confusion matrix 

```python
from sklearn.metrics import confusion_matrix 

a = confusion_matrix(y_test , y_predict )
```

```python
sns.heatmap(a , annot = True) ; 
```

```python
from sklearn.metrics import f1_score , precision_score , accuracy_score , recall_score
```

```python

print(f"{f1_score(y_test , y_predict )=}\n{precision_score(y_test , y_predict )=}\n{accuracy_score(y_test , y_predict )=}")
```

```python
from sklearn.metrics import classification_report 
classification_report(y_test , y_predict )
```

### K qiymat

```python
f1 = []
for k in range(1,15):
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(X_train , y_train )
    y_predict = knn.predict(X_test)
    f1.append(f1_score (y_predict , y_test ))
```

```python
plt.figure(figsize = (12,6))
plt.plot(range(1,15) , f1)
plt.grid()
plt.xticks(range(1,15));
```
